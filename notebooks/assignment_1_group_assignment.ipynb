```python
# column_4 from "leadTime" to "lead_time"
# column_6 from "Manufacturing Site" to "manufacturing_site"
# column_9 from "Order-Quantity" to "order_quantity"

def clean_data(data_1):
    data_1 = data_1.rename(columns={'leadTime': 'lead_time'})
    data_1 = data_1.rename(columns={'Manufacturing Site': 'manufacturing_site'})
    data_1 = data_1.rename(columns={'Order-Quantity': 'order_quantity'})
    return data_1

data_1_clean = clean_data(data_1)
data_1_clean.head()
```
